```python
from sklearn.model_selection import train_test_split

# Load the main dataset
dataset_path = '1.9MDataset.csv'
df = pd.read_csv(dataset_path)

# Sample approximately 2,500 entries
sample_size = 2500
sampled_df = df.sample(n=sample_size, random_state=42)  # Random sample

# Split the sampled dataset into training and testing datasets
train_ratio = 0.8  # 80% for training, 20% for testing
train_df, test_df = train_test_split(sampled_df, test_size=(1 - train_ratio), random_state=42)

# Save the datasets as CSV files
sampled_filename = 'main_sampled_dataset.csv'
train_filename = 'main_train_dataset.csv'
test_filename = 'main_test_dataset.csv'

sampled_df.to_csv(sampled_filename, index=False)
train_df.to_csv(train_filename, index=False)
test_df.to_csv(test_filename, index=False)

# Print confirmation
print(f"Main sampled dataset saved as '{sampled_filename}' ({len(sampled_df)} entries).")
print(f"Training dataset saved as '{train_filename}' ({len(train_df)} entries).")
print(f"Testing dataset saved as '{test_filename}' ({len(test_df)} entries).")
```

```python
genres = ['bakery', 'drinks', 'meal']

sample_size = 2500
train_ratio = 0.8  # 80% for training, 20% for testing

for genre in genres:
    # Load the genre-specific dataset
    genre_file = f'{genre}_dataset.csv'
    genre_df = pd.read_csv(genre_file)
    
    # Create a random sample of 2500 entries
    sampled_df = genre_df.sample(n=sample_size, random_state=42)
    
    # Split into training and testing datasets
    train_df, test_df = train_test_split(sampled_df, test_size=(1 - train_ratio), random_state=42)
    
    # Save the datasets
    sampled_filename = f'{genre}_sampled_dataset.csv'
    train_filename = f'{genre}_train_dataset.csv'
    test_filename = f'{genre}_test_dataset.csv'
    
    sampled_df.to_csv(sampled_filename, index=False)
    train_df.to_csv(train_filename, index=False)
    test_df.to_csv(test_filename, index=False)
    
    # Print confirmation
    print(f"Genre: {genre}")
    print(f"  - Sampled dataset: {sampled_filename} ({len(sampled_df)} entries)")
    print(f"  - Training dataset: {train_filename} ({len(train_df)} entries)")
    print(f"  - Testing dataset: {test_filename} ({len(test_df)} entries)")
    print()
```
